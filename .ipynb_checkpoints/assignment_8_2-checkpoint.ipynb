```python
#loading data
full_df = pd.read_csv("DOTS.csv")
full_df.head()
```

```python
#drop unneccesary columns
full_df = full_df.drop(['Country Code','Indicator Name','Indicator Code','Counterpart Country Code','Attribute','Unnamed: 8'], axis=1)
full_df.head()
```

```python
#replace null values with latest values and remove non integer 
full_df['2024M2'] = pd.to_numeric(full_df['2024M2'].astype(str).str.replace(',',''), errors='coerce').ffill().astype(int)
full_df.info()
```

```python
# making nodes

df = full_df.stack().reset_index()


g = nx.from_pandas_edgelist(full_df, 'Country Name', 'Counterpart Country Name', ['2024M2'])


nx.draw_spring(g, with_labels=True)
plt.show()
nx.draw_random(g, with_labels=True)
plt.show()
nx.draw_circular(g, with_labels=True)
plt.show()

```
